Visualization - WHO
===================

```r
WHO <- read.csv("WHO.csv")
```

```r
summary(WHO)
str(WHO)
```

In week one, the very first plot we made in R
was a scatterplot of fertility rate
versus gross national income.

Let's make this plot again, just like we did in week one.

```r
plot(WHO$GNI, WHO$FertilityRate)
```

Now, let's redo this scatterplot,
but this time using ggplot.

```r
library(ggplot2)
```

We need at least three things
to create a plot using ggplot

- data,
- aesthetic mapping o variables in the data frame to visual output,
- and a geometric object.

```r
scatterplot <- ggplot(WHO, aes(x=GNI, y=FertilityRate))
```

```r
scatterplot + geom_point()
```

```r
scatterplot + geom_line()
```

```r
scatterplot + geom_point(color="blue", size=3, shape=17)
```

```r
scatterplot + geom_point(color="darkred", size=3, shape=8)
```

```r
scatterplot + geom_point(color="darkred", size=3, shape=8) + ggtitle("Fertility Rate vs. Gross National Income")
```

```r
fertilityGNIplot <- scatterplot + geom_point(color="darkred", size=3, shape=8) + ggtitle("Fertility Rate vs. Gross National Income")
```

```r
pdf("MyPlot.pdf")
print(fertilityGNIplot)
dev.off()
```

```r
scatterplot + geom_point(color="darkred", size=3, shape=15) + ggtitle("Fertility Rate vs. Gross National Income")
```

```r
ggplot(WHO, aes(x=GNI, y=FertilityRate, color=Region)) + geom_point()
```

```r
ggplot(WHO, aes(x=GNI, y=FertilityRate, color=LifeExpectancy)) + geom_point()
```

```r
ggplot(WHO, aes(x=FertilityRate, y=Under15)) + geom_point()
```

```r
ggplot(WHO, aes(x=log(FertilityRate), y=Under15)) + geom_point()
```

```r
model <- lm(Under15 ~ log(FertilityRate), data=WHO)
summary(model)
```

```r
ggplot(WHO, aes(x=log(FertilityRate), y=Under15)) + geom_point() + stat_smooth(method="lm")
```

```r
ggplot(WHO, aes(x=log(FertilityRate), y=Under15)) + geom_point() + stat_smooth(method="lm", level=.99)
```

```r
ggplot(WHO, aes(x=log(FertilityRate), y=Under15)) + geom_point() + stat_smooth(method="lm", se=FALSE)
```

```r
ggplot(WHO, aes(x=log(FertilityRate), y=Under15)) + geom_point() + stat_smooth(method="lm", se=FALSE, color="orange")
```

```r
ggplot(WHO, aes(x = FertilityRate, y = Under15, color=Region)) + geom_point() + scale_color_brewer(palette="Dark2") 
```
